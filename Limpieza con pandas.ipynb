```python
import pandas as pd

df = pd.read_csv('Per Capita GDP of All Countries 1970 to 2022.csv')
print("Primeros registros del dataset:")
display(df.head())

print("\nInformación general del dataset:")
df.info()

print("\nDatos nulos por columna:")
print(df.isnull().sum())

df = df.dropna()
df = df.drop_duplicates()
print("\nDatos después de la limpieza:")
display(df.head())

df.to_csv('Per Capita GDP of All Countries 1970 to 2022_cleaned.csv', index=False)
```
